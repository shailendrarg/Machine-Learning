```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model # for machine learning REGRESSION
from scipy.interpolate import *

df= pd.read_csv("House_file.csv")
df
```

```python
#dropping last columns
df.dropna(inplace=True,axis=1)
```

```python
df.isnull().sum() #check for missing data
```

```python
#check for duplicate data
df.duplicated().sum()
```

```python
#setting Home as index
df.set_index("Home",inplace = True)
```

```python
#Plotting the data as per the matplotlib
plt.scatter(df.SqFt,df.Price,visible=True)
```

```python
#linearr regression
reg=linear_model.LinearRegression()
reg
```

```python
reg.fit(df[["SqFt"]],df.Price)# fit(2d array,y intercept)
```

```python
reg.coef_
```

```python
reg.intercept_
```
