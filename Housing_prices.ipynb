```python
missing_data=df.isna().sum()/len(df)*100
missing_data
```

```python
#remove columns having 80% of missing data
remove_col=[]
for data in missing_data.index:
    if missing_data[data]>80.0:
        remove_col.append(data)
print(remove_col)    
```

```python
df.drop(columns=remove_col,inplace=True,axis=1)
```

```python
df.shape
```

```python
cat_col=[]
num_col=[]
cat_count=0
num_count=0
for col in df.columns:
    if df[col].dtype=="object":
        cat_col.append(col)
        cat_count=cat_count+1
    else:
        num_col.append(col)
        num_count=num_count+1
    
        
```

```python
print(cat_col,cat_count)
print(num_col,num_count)
```
