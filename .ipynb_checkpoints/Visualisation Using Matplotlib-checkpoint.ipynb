# Matlplot lib tutorial


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Plotting the x,y in a line
x=[1,2,3,4,5,6]
y=[10,80,30,45,50,60]
plt.xlabel("Numbers ---------->")
plt.ylabel("Qunatitys ------------->")
plt.title("Numbers and Quantity")
plt.plot(x,y,color='g',linewidth=2,linestyle="--",marker="*",markerfacecolor="b",visible=True,antialiased=True,alpha=0.8)
#alpha ranges from 0 to 1, it makes it visible and invisble.
```

```python
#plotting different figures in matplotlib spaces
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])
```

# Histogram

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])  #ranges of x and y. min x is 40 and max is 160, min of y is 0 and max is 0.03 
plt.grid(True)
plt.show()
```

# Axes,labels,Legends and Grid

```python
#Plotting the x,y in a line
x=[1,2,3,4,5,6]
y=[10,80,30,45,50,60]
a=[11,22,33,44,55,66]
b=[10,80,30,45,50,60]
plt.grid(True)
plt.xlabel("Numbers ---------->")
plt.ylabel("Qunatitys ------------->")
plt.title("Numbers and Quantity")
plt.plot(x,y,color='g',linewidth=2,label="Gamma")
plt.plot(a,x,label="alpha")
plt.plot(a,y,label="Beta")
plt.legend(loc="best",shadow =True)# it highlights the legend corners)
#default for position of legend is best, other properties are: upper left/right

```

# Bargraph: Used in business like,revenues,profits,hikes,sales etc

```python
#new example
Company =["Google","Microsoft","Apple","Amazon"]
profits =[100,120,200,300]
sales=[120,140,150,160]
plt.figure(1)
plt.xlabel("Company")
plt.ylabel("Sales")
plt.title("Company and sales report")
plt.bar(Company,sales,color="g",width=0.1)
plt.show()
plt.figure(2)
plt.xlabel("Company")
plt.ylabel("Profits")
plt.title("Company and Profits report")
plt.bar(Company,profits,color="r",width=0.4)
plt.show()
```

# Doing the same in one diagram

```python
Company =["Google","Microsoft","Apple","Amazon"]
profits =[100,120,200,300]
sales=[120,140,150,160]
plt.xlabel("Company")
plt.ylabel("Sales")
plt.title("Company and sales report")
plt.bar(Company,profits,color="r",width=0.4)
plt.bar(Company,sales,color="g",width=0.1)
plt.show()
```

```python
#Pie chart learning
plt.pie(sales,labels=Company,radius=2,autopct="%0.2f%%",shadow=True,explode=[0.2,0.3,0.4,0.2],startangle =45)
plt.axis("equal")
plt.show()
```
