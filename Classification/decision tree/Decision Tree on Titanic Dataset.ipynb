```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.columns
```

```python
titanic=df[['Survived', 'Pclass','Sex','Age','Fare']]
titanic.head()
```

```python
titanic.isnull().sum() #few values in age are not known
```

```python
#since sex is a categorical variable, we need to do encoding for the same

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Sex1=le.fit_transform(titanic.Sex)

titanic["Sex1"]=Sex1

titanic
```

```python
titanic.drop(columns=["Sex","Age"],axis=1,inplace=True)
#Age is dropped as it had null value and cannot be used, either fillna or dropna can be used.
titanic.rename(columns={"Sex1":"Sex"})
```

```python
X=titanic.drop(columns="Survived",axis="columns")
Y=titanic.Survived
```

```python
#lets train test and split the data for descision tree
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
model.fit(X_train,Y_train) # training
```

```python
model.score(X_test,Y_test) # model performance
```

#plotting the decision Tree

```python
from sklearn.tree import export_graphviz
export_graphviz(model,feature_names=["Plcass","Age","Sex"],out_file="decision_Tree.jpeg",class_names="Survived",
    label='all',
    filled=False,
    leaves_parallel=False,
    impurity=True,
    node_ids=False,
    proportion=False)
```
